```python
import pandas as pd
df =pd.read_csv('/content/drive/MyDrive/datasets/wholesale.csv')
df.head()
```

```python
df =df.drop('Series_reference',axis ='columns')
df =df.drop('Suppressed',axis ='columns')
df =df.drop('Group',axis ='columns')
df =df.drop('Unnamed: 13',axis ='columns')

df.head()
```

```python
inputs=df.drop('Series_title_4',axis='columns')

target = df['Series_title_4']
```

```python
inputs.head()
```

```python
import math
median_Period = math.floor(inputs.Period.median())
median_Data_value = math.floor(inputs.Data_value.median())
median_Period
median_Data_value
```

```python
inputs.Period = inputs.Period.fillna(median_Period)
inputs.Data_value = inputs.Data_value.fillna(median_Data_value)
```

```python
target.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le_STATUS	=LabelEncoder()
le_UNITS	=LabelEncoder()
le_Subject	=LabelEncoder()
le_Series_title_1=LabelEncoder()
le_Series_title_2	=LabelEncoder()
le_Series_title_3=LabelEncoder()
#le_Series_title_4=LabelEncoder()

```

```python
inputs['STATUS'] = le_STATUS.fit_transform(inputs['STATUS'])
inputs['UNITS'] = le_UNITS.fit_transform(inputs['UNITS'])
inputs['Subject'] = le_Subject.fit_transform(inputs['Subject'])
inputs['Series_title_1'] = le_Series_title_1.fit_transform(inputs['Series_title_1'])
inputs['Series_title_2'] = le_Series_title_2.fit_transform(inputs['Series_title_2'])
inputs['Series_title_3'] = le_Series_title_3.fit_transform(inputs['Series_title_3'])
#inputs['Series_title_4'] = le_Series_title_4.fit_transform(target['Series_title_4'])

```

```python
inputs.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(inputs,target)
```

```python
x_train.head()
```

```python
from sklearn import tree
```

```python
tree = tree.DecisionTreeClassifier()
```

```python
tree.fit(x_train,y_train)
```

```python
tree.predict(x_test)
```

```python

tree.score(x_test,y_test)
```

```python
knn.score(x_test,y_test)
```
